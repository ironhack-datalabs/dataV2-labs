#### Import the necessary libraries here:

```python
# your code here
from stackapi import StackAPI
from datetime import datetime
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
SITE = StackAPI('stackoverflow')
SITE.page_size = 100
SITE.max_pages = 1  ## Did made this one 100 once and it gave 10000 results, so there are even more....
# questions = SITE.fetch('questions', fromdate=1585699200, todate=1588291200)
# questions

answers = SITE.fetch('answers', fromdate=1585699200, todate=1588291200)
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
questions_most_voted = SITE.fetch('questions', fromdate=datetime(2020,5,20), todate=datetime(2020,5,21),
    min = 5, tagged='python', sort = 'votes')
questions_most_voted
```
