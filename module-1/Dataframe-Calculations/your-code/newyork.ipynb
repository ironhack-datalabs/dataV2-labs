```python
df = df[df['Work Location Borough'] != 'Unknown']
```

```python
import string
def remove_punctuations(words):
    buffer = []
    punc = string.punctuation
    for w in words:
        if w not in punc:
            buffer.append(w)
        else:
            continue
    return ''.join(buffer).lstrip()
```

```python
df['Title Description'] = df['Title Description'].apply(remove_punctuations)
```
