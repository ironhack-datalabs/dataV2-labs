## Création d'une liste pour concaténer toutes les données offres d'alternance des différentes zones françaises

```python
df_list = []

df_list.append(json_normalize(data_centre['peJobs']['results']))

df_list.append(json_normalize(data_nord_est['peJobs']['results']))

df_list.append(json_normalize(data_nord_ouest['peJobs']['results']))

df_list.append(json_normalize(data_sud_est['peJobs']['results']))

df_list.append(json_normalize(data_nord['peJobs']['results']))

df_list.append(json_normalize(data_sud_ouest['peJobs']['results']))

jobs_df = pd.concat(df_list, axis=0)

jobs_df
```

## Création des fonctions pour obtenir les n° de départements + les mois concernés

```python
#Ajout d'une colonne pour n'obtenir que le n° du département

def get_department(x):
    return((x[:2]))

jobs_df['Dpt'] = jobs_df['place.zipCode'].apply(str).apply(get_department)
```

```python
def get_month(x):
    return(x[0:7])

jobs_df['month'] = jobs_df['job.creationDate'].apply(get_month)

filtered_jobs_df = jobs_df[['job.creationDate','month','title','place.city','place.zipCode','Dpt','company.name']]

filtered_jobs_df
```

## Ouverture & Lecture des données 

```python
data_covid = pd.read_csv('table-indicateurs-open-data-dep-2021-04-09-20h00.csv')

covid_df = pd.DataFrame(data_covid)

#covid_df a initialement 39188 entrées
```

## Obtenir les mois en fonction de la date d'analyse

```python
covid_df['month'] = covid_df['date'].apply(get_month)
```

## Transformer les dates dans le bon format dans le but de pouvoir appliquer un filtre

```python
covid_df['date'] = pd.to_datetime(covid_df['date'], format='%Y-%m-%d')
```

```python
covid_df[['dep', 'date', 'month', 'tx_incid']]
```

## Création d'un filtre pour n'obtenir les dates qu'à partir du 01-08-2021

```python
mask = (covid_df['date'] > '2020-08-1') & (covid_df['date'] <= '2021-04-11')
filtered_covid_df=covid_df.loc[mask]
data_covid_df = filtered_covid_df[['dep', 'date', 'month', 'tx_incid']]
data_covid_df

#filtered_covid_df a désormais 25351 entrées
```

# Définition du taux d'incidence :

1. Activité épidémique (taux d'incidence)

Le taux d'incidence correspond au nombre de tests virologiques positifs pour 100.000 habitants sur une semaine. Il est calculé, à l'échelle départementale, de la manière suivante : (100000*nombre de cas positif)/population.

Trois niveaux ont été fixés pour cet indicateur :

vert : en dessous de 10 personnes testées positives sur 100.000 personnes testées, sur une semaine glissante ;
orange : au-delà de 10 personnes testées positives sur 100.000 personnes testées, sur une semaine glissante ;
rouge : au-delà de 50 personnes testées positives sur 100.000 personnes testées, sur une semaine glissante.
