## Challenge 1: Fork Languages

You will find out how many programming languages are used among all the forks created from the main lab repo of your bootcamp.

`````python
# import libraries here
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize

username = 'XXXX'
token = 'XXXX'
`````

Assuming the main lab repo is ironhack-datalabs/madrid-oct-2018, you will:

#### 1. Obtain the full list of forks created from the main lab repo via Github API.

To list forks, we can use the GET method. As explained in the GitHub API documentation, we need to make the request to: GET /repos/:owner/:repo/forks.

`````python
# # your code here

response_1 = requests.get('https://api.github.com/repos/ironhack-datalabs/dataV2-labs/forks?page=1&per_page=100', auth=(username,token))
results_1 = response_1.json()
data_1 = pd.DataFrame(results_1)

response_2 = requests.get('https://api.github.com/repos/ironhack-datalabs/dataV2-labs/forks?page=2&per_page=100', auth=(username,token))
results_2 = response_2.json()
data_2 = pd.DataFrame(results_2)

data_complete_forks = pd.concat([data_1, data_2])
data_complete_forks
`````

`````python
sublist = []
i = 1
response = requests.get('https://api.github.com/repos/ironhack-datalabs/dataV2-labs/forks?page=' + str(i) + '&per_page=100', auth=(username,token))
response_json = response.json()
#While there is something in the json object returned, keep increasing the page number
while len(response_json)>0:
    #Make the request
    response = requests.get('https://api.github.com/repos/ironhack-datalabs/dataV2-labs/forks?page=' + str(i) + '&per_page=100', auth=(username,token))
    #Print the result to make sure is successful
    print(response)
    #Assign the returned object to a variable
    response_json = response.json()
    #Append this variable in my list
    sublist.append(response_json)
    #Increase the page
    i += 1
    print(len(response_json))


print(len(sublist)) 
#Convert the list of list of json objects in a single list of json objects
flat_list = [item for element in sublist for item in element]
print(len(flat_list))
data_total = pd.DataFrame(flat_list)
data_total
`````

## Challenge 2: Count Commits
Count how many commits were made in the month of october of 2018.
#### 1. Obtain all the commits made in October 2018 via API, which is a JSON array that contains multiple commit objects.

`````python
# your code here

response_3 = requests.get('https://api.github.com/repos/ironhack-datalabs/dataV2-labs/commits?page=1&per_page=100', auth=(username,token))
results_3 = response_3.json()
flattened_data_3 = json_normalize(results_3)

response_4 = requests.get('https://api.github.com/repos/ironhack-datalabs/dataV2-labs/commits?page=2&per_page=100', auth=(username,token))
results_4 = response_4.json()
flattened_data_4 = json_normalize(results_4)

data_complete_commits = pd.concat([flattened_data_3, flattened_data_4])
data_complete_commits
`````

#### 2. Count how many commit objects are contained in the array.

`````python
# your code here


# data_complete_commits.columns

check_date = data_complete_commits['commit.author.date']

yd_list = [i[:7] for i in list(check_date) if i[:7] == '2019-10']
print(f'The number of commits made in October 2019 is: {len(yd_list)}.')

`````
