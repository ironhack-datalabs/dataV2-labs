# Data Cleaning 

#### 1. Import pandas library.

```python
import pandas as pd
import numpy as np
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
import pymysql
from sqlalchemy import create_engine
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
#hostname: relational.fit.cvut.cz
#port: 3306
#username: guest
#password: relational
my_engine = create_engine('mysql+pymysql://guest:relational@relational.fit.cvut.cz/stats')

```

#### 4. Import the users table.

```python
user_table = pd.read_sql_query('SELECT * FROM stats.users',my_engine)
user_table.head()
```

#### 5. Rename Id column to userId.

```python
change_ID = user_table.rename(columns={'Id':'userId'})
change_ID.head()
#df_new = df.rename(columns={'A': 'a'}, index={'ONE': 'one'})
```

#### 6. Import the posts table. 

```python
posts_table = pd.read_sql_query('SELECT * FROM stats.posts',my_engine)
posts_table.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
ID_change = posts_table.rename(columns={'Id': 'postId', 'OwnerUserId':'userId'})
ID_change.head()
                      
                 
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
users_df = change_ID[['userId','Reputation','Views','UpVotes','DownVotes']]
posts_df = ID_change[['postId','Score','userId','ViewCount','CommentCount']]

print(users_df.shape)
print(posts_df.shape)

```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
#merging the 2 dataframes on userId

users_posts = pd.merge(users_df,posts_df, how ='inner', on='userId')

users_posts




```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

```python
#users_posts.isnull()
#finding columns with null values
null_cols = users_posts.isnull().sum()

print (null_cols)


print('More than 1/2 of the viewcounts are empty')
```

#### 11. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before going to the next step.

```python
print('still having 1/2 of the information can yield some results.  I will fill them to keep the data.')
print('\n')
null_cols1=users_posts.replace(np.nan,'Unknown',regex=True)

print (null_cols1)
```
