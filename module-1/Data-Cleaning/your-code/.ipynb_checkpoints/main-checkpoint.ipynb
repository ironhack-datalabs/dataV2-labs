# Data Cleaning 

#### 1. Import pandas library.

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
import pymysql
from sqlalchemy import create_engine
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
engine = create_engine('mysql+pymysql://guest:relational@relational.fit.cvut.cz/stats')
```

#### 4. Import the users table.

```python
users = pd.read_sql_query('SELECT * FROM stats.users;', engine)
users.head()
```

#### 5. Rename Id column to userId.

```python
users = users.rename(columns = {'Id':'userId'})
users.head()
```

#### 6. Import the posts table. 

```python
posts = pd.read_sql_query('SELECT * FROM stats.posts;', engine)
posts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts = posts.rename(columns = {'Id':'postId', 'OwnerUserId':'userId'})
posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
users_columns = ['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']
posts_columns = ['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']

users1 = users[users_columns]
posts1 = posts[posts_columns]
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
# Inner merge
merged_data = users1.merge(posts1, how='inner')
merged_data.head()
```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

```python
null_cols = merged_data.isnull().sum()
null_cols

# Tehre are missing values in ViewCount column
```

#### 11. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before going to the next step.

```python
print("merged_data shape:", merged_data.shape)

# I do not want to drop the column ViewCount, because slightly less than half of it has meaningful values.

print(merged_data['ViewCount'].describe())

# It could replace ViewCount NaN values with 0 (zero post views)

merged_data[['ViewCount']] = merged_data[['ViewCount']].fillna(0)
merged_data.head()
```

#### 12. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
# ViewCount as an integer would make more sense (you cannot have 1,5 views).
# Moreover, original dataset has it as an integer, i.e. the type changed after NaN replacement.

print(merged_data.dtypes)

merged_data['ViewCount'] = merged_data['ViewCount'].astype('int64')

#  Doublechecking if the type change was successful

print(merged_data.dtypes)
```
