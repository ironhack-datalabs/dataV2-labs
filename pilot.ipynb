```python
sharks_clean['Date2'] = sharks_clean['Date']
def cast_date(date):
    date = str(date).strip()
    date = date[:10].replace('.','-')
    if date.endswith('00'):
        date = list(date)
        date[-1] = '1'
        return  "".join(date)
    return date
    

sharks_clean['Date'] = sharks_clean['Case_Number'].apply(cast_date)

#Creamos un Date2 solo para preservar informacion importante ya que Date seria remplazada con otro formato
#A partir de Case_Number obtuvimos valores correctos de Date remplazando . por -
```

```python
index_list = list(sharks_clean[sharks_clean['Case_Number'].str.contains('ND', na=False)].index)


for i in index_list:
    if any(chr.isdigit() for chr in sharks_clean['Date2'][i] )== False:
        sharks_clean['Date2'][i] = None
    
    else:   
         sharks_clean['Date2'][i] = re.findall(r'[0-9]{4}',sharks_clean['Date2'][i])[0]
        
#Existian regsitros en Case_Number que no tenian la fecha completa y comenzaban con ND obtuvimos una lista con los index
#Estos registros solo contenian el aÃ±o; en el ciclo for obtuvimos el aÃ±o contenido en los registros desordenados de Case_Number
#Lo colocamos en Date2

```
