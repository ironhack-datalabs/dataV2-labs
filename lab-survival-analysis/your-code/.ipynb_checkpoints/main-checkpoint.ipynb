## 1. Generate and plot a survival function that shows how employee retention rates vary by gender and employee age.

*Tip: If your lines have gaps in them, you can fill them in by using the `fillna(method=ffill)` and the `fillna(method=bfill)` methods and then taking the average. We have provided you with a revised survival function below that you can use for the exercises in this lab*

```python
def survival(data, group_field, time_field, event_field):
    kmf = KaplanMeierFitter()
    results = []

    for i in data[group_field].unique():
        group = data[data[group_field]==i]
        T = group[time_field]
        E = group[event_field]
        kmf.fit(T, E, label=str(i))
        results.append(kmf.survival_function_)

    survival = pd.concat(results, axis=1)
    front_fill = survival.fillna(method='ffill')
    back_fill = survival.fillna(method='bfill')
    smoothed = (front_fill + back_fill) / 2
    return smoothed
```

```python
# Note: timeline includes Age 0 (need to check why that is). We have to disregard 
# it now, because working age starts at 18.

rates = survival(data, 'Gender', 'Age', 'Attrition')
rates
```

```python
rates.iplot(kind='line', xTitle='Employee age', yTitle='Retention Rate',
            title='Retention Rates by Employee age and Gender')
```

## 9. Let examine how compensation affects attrition.

- Use the `pd.qcut` method to bin the HourlyRate field into 5 different pay grade categories (Very Low, Low, Moderate, High, and Very High).
- Generate and plot survival curves showing employee retention by pay grade and age.

```python
data['PayGrade'] = pd.qcut(data.HourlyRate, 5, labels=["Very Low", "Low", "Moderate", "High", "Very High"])

rates = survival(data, 'PayGrade', 'Age', 'Attrition')

rates.iplot(kind='line', xTitle='Age', yTitle='Retention Rate',
            title='Retention Rates by Pay grade and Employee age')
```

## 10. Finally, let's take a look at how the demands of the job impact employee attrition.

- Create a new field whose values are 'Overtime' or 'Regular Hours' depending on whether there is a Yes or a No in the OverTime field.
- Create a new field that concatenates that field with the BusinessTravel field.
- Generate and plot survival curves showing employee retention based on these conditions and employee age.

```python
data.loc[data['OverTime'] == 'Yes', 'TimeDemands'] = 'Overtime'  
data.loc[data['OverTime'] != 'Yes', 'TimeDemands'] = 'Regular Hours'

data['Travel_Overtime'] = data['TimeDemands']+" & "+data['BusinessTravel']

rates = survival(data, 'Travel_Overtime', 'Age', 'Attrition')

rates.iplot(kind='line', xTitle='Age', yTitle='Retention Rate',
            title='Retention Rates by Job demands and Employee age')
```
