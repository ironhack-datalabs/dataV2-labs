```python
import datetime
import re

year_first = re.compile("\d{4}")

def convertDate(date):
    if year_first.match(date):
        return datetime.datetime.strptime(date, "%Y-%m-%d %H:%M:%S")
    else:
        return datetime.datetime.strptime(date, "%m/%d/%y %H:%M")
    
def extractYearAndMonth(date):
    year = str(date.year)
    month = date.month
    if month < 10:
        month = '0' + str(month)
    else:
        month = str(month)
    return year + '-' + month

def extractYear(date):
    return str(date.year)

def extractQuarter(date):
    quarter = str(date.year) + 'Q'
    if date.month < 4:
        quarter += '1'
    elif date.month < 7:
        quarter += '2'
    elif date.month < 10:
        quarter += '3'
    else:
        quarter += '4'
    return quarter

twitter['created_datetime'] = twitter['created_at'].apply(lambda date: convertDate(date))
twitter['year_month'] = twitter['created_datetime'].apply(lambda date: extractYearAndMonth(date))
twitter['year'] = twitter['created_datetime'].apply(lambda date: extractYear(date))
twitter['quarter'] = twitter['created_datetime'].apply(lambda date: extractQuarter(date))

twitter['absolute_reach'] = twitter['favourite_count'] / 2 + twitter['retweet_count']
twitter.head()
```

```python
twitter_hist = pd.read_csv('../data/twitter_history.csv')
twitter_hist.drop(columns=['Name'], inplace=True)
twitter_hist.set_index(['twitter_name'], inplace=True)
twitter_hist
```

```python
estimated_followers=[]
relative_reach=[]
for index,row in twitter.iterrows():
    estimated_followers_tweet = twitter_hist.loc[row['screen_name'], row['quarter']]
    try:
        estimated_followers_tweet = int(estimated_followers_tweet.replace(",",""))
        relative_reach_tweet = 1000000 * row['absolute_reach'] / estimated_followers_tweet
    except:
        estimated_followers_tweet = float('NaN')
        relative_reach_tweet = float('NaN')

    estimated_followers.append(estimated_followers_tweet)
    relative_reach.append(relative_reach_tweet)

twitter['estimated_followers']=estimated_followers
twitter['relative_reach']=relative_reach

twitter
```
