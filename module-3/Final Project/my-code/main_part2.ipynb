```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.stem.snowball import EnglishStemmer
import random
```

```python
twitter_mod = pd.read_csv('../data/twitter_modified.csv')
twitter_mod.reset_index(drop=True, inplace=True)
twitter_mod.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
twitter_mod.to_csv('twitter_modified.csv')
twitter_mod
```

```python
def clean_up(s):
    #return re.sub("[\W\d]+"," ", re.sub("http://\S*", "", s)).lower()
    s_1 = re.sub('\d', '', s)
    s_2 = re.sub('(http|https)://([^\s]+)', '', s_1)
    s_3 = re.sub('\W', ' ', s_2 )
    s_4 = (s_3.lower())
    return s_4

def tokenize(s):
    return nltk.word_tokenize(s)

def stem_and_lemmatize(l):
    lemmatizer = WordNetLemmatizer()
    stem = EnglishStemmer()
    return [stem.stem(lemmatizer.lemmatize(i)) for i in l]

def remove_stopwords(l):
    filtered_sentence = []
    for r in l: 
        if not r in stopwords.words('english'):
            filtered_sentence.append(r)
    return filtered_sentence
```

```python
twitter_mod['text_processed'] = twitter_mod.apply(lambda x: clean_up(x['text']), axis=1)
twitter_mod['text_processed'] = twitter_mod.apply(lambda x: tokenize(x['text_processed']), axis=1)
twitter_mod['text_processed'] = twitter_mod.apply(lambda x: stem_and_lemmatize(x['text_processed']), axis=1)
twitter_mod['text_processed'] = twitter_mod.apply(lambda x: remove_stopwords(x['text_processed']), axis=1)
```

```python
import time

def clean_words (row):
    return row.apply(lambda row: [x for x in row if x not in ['amp', 'http']])

words = {}
for word_list in twitter_mod['text_processed']:
    for word in word_list:
        if word not in words:
            words [word] = {}

def calculate_tfidf (tweets, accountName):
    print('Calculating ' + accountName + '...')
    start = time.time()
    cutoff = 5000
    if (accountName != 'ALL'):
        # If accountName is not ALL, we only select the tweets of a specific account
        tweets = tweets[tweets['screen_name'] == accountName]
    
    # We filter tweets into two categories, above and below the relative reach cutoff value
    above = clean_words(tweets[(tweets['relative_reach'] >= cutoff)])
    below = clean_words(tweets[(tweets['relative_reach'] < cutoff)])
    
    # We calculate tf for the words on both groups and we calculate idf for all the words
    above_tf = FreqDist()
    below_tf = FreqDist()
    idf      = FreqDist()

    for word_list in above.text_processed:
        for word in word_list:
            above_tf[word] +=1
    for word_list in below.text_processed:
        for word in word_list:
            below_tf[word] +=1
    for word_list in tweets.text_processed:
        for word in word_list:
            idf[word] +=1
    for key in idf:
        idf[key]=math.log(tweets['text_processed'].count () / idf[key])
    
    # We build the words object with the words and the scores per word
    for word in words:
        if word in above_tf:
            above_tfidf = above_tf[word]*idf[word]
        else:
            above_tfidf = 0
        if word in below_tf:
            below_tfidf = below_tf[word]*idf[word]
        else:
            below_tfidf = 0
            
        words[word][accountName + '-above_tfidf'] = above_tfidf
        words[word][accountName + '-below_tfidf'] = below_tfidf
        if below_tfidf == 0:
            if above_tfidf > 0:
                # if a word is on the above group but not on the below, we give it a "high score" of 10
                words[word][accountName + '-ratio_tfidf'] = 10
            else:
                words[word][accountName + '-ratio_tfidf'] = 0
        else:
            words[word][accountName + '-ratio_tfidf'] = above_tfidf/below_tfidf
    
    print('Done with ' + accountName + ' in ' + str(time.time() - start))

accounts = ['4ocean', '5gyres', 'aplastic_planet', 'DebrisTracker',
       'globalalert', 'JackieSavitz', 'LifeWoutPlastic', 'merijntinga',
       'MissionBlue', 'NewPlasticsEcon', 'oceana', 'oceanfdn',
       'oceanunite', 'Oceanwarrior', 'OurOcean', 'PlasticBank',
       'PlasticFreeJuly', 'PlasticOceans', 'PlasticOceansUS',
       'plasticpdise', 'PlasticPollutes', 'PlasticsBeyond',
       'plasticsoupfoun', 'RethinkPlastic', 'sascampaigns', 'SylviaEarle',
       'TBPInnovate', 'TheOceanCleanup', 'ThePlasticWhale', 'ALL']

for account in accounts:
    calculate_tfidf(twitter_mod, account)
    
words = pd.DataFrame.from_dict(words, orient='index')

words
```
