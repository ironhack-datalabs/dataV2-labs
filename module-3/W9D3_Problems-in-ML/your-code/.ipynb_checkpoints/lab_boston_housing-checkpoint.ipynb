## Developing a Model

### Implementation: Define a Performance Metric
What is the performance meteric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    # Your code here:
    return r2_score(y_true, y_predict)
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
# Your code here
from sklearn.model_selection import train_test_split

X = data.drop(['medv'], axis=1)
y = data['medv']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # shuffling is default
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
# Five separate RFR here with the given max depths
from sklearn.ensemble import RandomForestRegressor

RFR_2 = RandomForestRegressor(max_depth=2)
RFR_4 = RandomForestRegressor(max_depth=4)
RFR_6 = RandomForestRegressor(max_depth=6)
RFR_8 = RandomForestRegressor(max_depth=8)
RFR_10 = RandomForestRegressor(max_depth=10)

RFR_2.fit(X_train, y_train)
y_pred_2_test = RFR_2.predict(X_test)
y_pred_2_train = RFR_2.predict(X_train)
RFR_4.fit(X_train, y_train)
y_pred_4_test = RFR_4.predict(X_test)
y_pred_4_train = RFR_4.predict(X_train)
RFR_6.fit(X_train, y_train)
y_pred_6_test = RFR_6.predict(X_test)
y_pred_6_train = RFR_6.predict(X_train)
RFR_8.fit(X_train, y_train)
y_pred_8_test = RFR_8.predict(X_test)
y_pred_8_train = RFR_8.predict(X_train)
RFR_10.fit(X_train, y_train)
y_pred_10_test = RFR_10.predict(X_test)
y_pred_10_train = RFR_10.predict(X_train)
```

Now, plot the score for each tree on the training set and on the testing set.

```python
# Produce a plot with the score for the testing and training for the different max depths
score_2_test = performance_metric(y_test, y_pred_2_test)
score_4_test= performance_metric(y_test, y_pred_4_test)
score_6_test = performance_metric(y_test, y_pred_6_test)
score_8_test = performance_metric(y_test, y_pred_8_test)
score_10_test = performance_metric(y_test, y_pred_10_test)
scores_test = [score_2_test, score_4_test, score_6_test, score_8_test, score_10_test]
print(scores_test)

score_2_train = performance_metric(y_train, y_pred_2_train)
score_4_train = performance_metric(y_train, y_pred_4_train)
score_6_train = performance_metric(y_train, y_pred_6_train)
score_8_train = performance_metric(y_train, y_pred_8_train)
score_10_train = performance_metric(y_train, y_pred_10_train)
scores_train = [score_2_train, score_4_train, score_6_train, score_8_train, score_10_train]
print(scores_train)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(range(2,11,2), scores_train, label = "train")
ax.plot(range(2,11,2), scores_test, label = "test")
ax.legend()
plt.show()
```
