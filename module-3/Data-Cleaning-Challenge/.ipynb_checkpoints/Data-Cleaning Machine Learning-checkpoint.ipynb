```python
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("iris-data.csv", index_col=0)
```

```python
iris.isna().sum()
```

```python
iris.dtypes
```

```python
iris["petal_width_cm"].mean()
```

```python
iris["class"].unique()
```

```python
iris[iris["class"] == "Iris-virginica"].isna().sum()                        
```

```python
my_list = []
for flower in iris["class"].unique(): 
    print(flower)
    temporary_df = iris[iris["class"] == flower]
    my_list.append(temporary_df.fillna(temporary_df["petal_width_cm"].mean()))
                   
```

```python
iris = pd.concat(my_list)
iris
```

```python
iris.plot.box()
```

```python
import seaborn as sns
import matplotlib as plt

sns.pairplot(iris, hue = 'class')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.drop(["class"],axis=1), iris["class"],random_state = 0, test_size = 0.20)
```

```python
y_test.shape
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn
```

```python
knn.fit(X_train, y_train)
```

```python
y_predict = knn.predict(X_test)
y_predict
```

```python
y_test
```

```python
knn.score(X_test, y_test)
```
