```python
iris["petal_width_cm"] = iris["petal_width_cm"].fillna(iris["petal_width_cm"][iris["class"]=="Iris-setosa"].mean())
```

```python
lista = []
for flower in iris["class"].unique(): 
    print(flower)
    temporary_df = iris[iris["class"] == flower]
    lista.append(temporary_df.fillna(temporary_df["petal_width_cm"].mean()))
```

```python
iris2 = pd.concat(lista)
iris2.head(15)
```
