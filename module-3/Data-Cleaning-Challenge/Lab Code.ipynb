```python
import pandas as pd
```

```python
iris = pd.read_csv('iris-data.csv')
iris
```

```python
##check the data is encoded/formatted properly
iris.dtypes

#they are all float which is what we wanted - so yes
```

```python
#Find any null values
iris.isna().sum()
```

```python
#we can see that petal_width_cm has null values which I have chosen to replace with the mean

#So finding the mean

iris["petal_width_cm"].mean()
```

```python
iris.describe()
```

```python


iris['petal_width_cm'].groupby(iris['class']).mean()
```

```python
#replacing null values with the mean of the class
iris['petal_width_cm'].fillna(iris['petal_width_cm'].groupby(iris['class']).mean())
```

```python

#check it's worked - null values are filled
iris.isna().sum()
```

```python
#We can see from the box plot that there are outliers/ they do not all fall within the expected mean? 
#I think pls confirm if this is correct or not?

iris.plot.box()
```
