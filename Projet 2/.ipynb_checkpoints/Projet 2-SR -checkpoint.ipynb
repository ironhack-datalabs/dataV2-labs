```python
#Convert "Année" column to numeric
```

```python
df["Année"] = pd.to_numeric(df["Année"])
```

```python
#remove "," from 3rd column
```

```python
df["Personnes décédées sur le coup ou dans les 30 jours après l'accident"] = df["Personnes décédées sur le coup ou dans les 30 jours après l'accident"].str.replace(',', '')

df
```

```python
df["Personnes décédées sur le coup ou dans les 30 jours après l'accident"] = pd.to_numeric(df["Personnes décédées sur le coup ou dans les 30 jours après l'accident"])
df["Milliards véh.km"] = pd.to_numeric(df["Milliards véh.km"])
df["T30j/milliard de véhxKm"] = pd.to_numeric(df["T30j/milliard de véhxKm"])


df.dtypes
```

```python
#Add growthrate in a new column
```

```python
df["Growth Rate"] = df["Personnes décédées sur le coup ou dans les 30 jours après l'accident"].pct_change(limit=2)*100

df
```

```python
#get second dataset from Wikipedia
```

```python
url = 'https://fr.wikipedia.org/wiki/S%C3%A9curit%C3%A9_routi%C3%A8re_en_France'

html = requests.get(url).content
html
```

```python
soup = BeautifulSoup(html, "html")
print(soup.prettify())
```

```python
#Get all li from page
```

```python
all_li = soup.find_all("li")

table_name_laws = []

for line in all_li:
    law = line.get_text()
    table_name_laws.append(law)
                                     
table_name_laws

table_name_laws = list(map(lambda st: str.replace(st, "\xa0", " "), table_name_laws))

table_name_laws



```

```python
#keep all "li" with the pattern "date : "
```

```python
r = re.compile('\d{4} :')
newlist = list(filter(r.match, table_name_laws))
print(newlist)

```

```python
#Separate "dates" from "description of law"
```

```python
laws = DataFrame(newlist,columns=['Lois'])

laws = laws.Lois.apply(lambda x: pd.Series(str(x).split(":")))

laws
```

```python
#Rename "header"
```

```python
new_names =  {0 : 'Année',
              1:"Lois1",
              2:"Lois2"
             }

laws.rename(columns=new_names, inplace=True)

laws
```

```python
#Convert "Année" to int
```

```python
laws["Année"] = laws["Année"].astype(int)

laws.dtypes
```

```python
#Get a third column which concat Lois1 and Lois2
```

```python
laws['Lois'] = laws['Lois1'] + laws['Lois2']


laws.dtypes
```

```python
for i in laws["Lois1"]:
    if laws["Lois2"] is not "NaN":
        laws["Lois"] = laws["Lois1"]
        
laws
```

```python
cols = ["Lois1","Lois2"]

laws = laws.drop(cols, axis=1)

laws
```

```python
df = df.loc[:, df.columns.notnull()]
```

```python
df = df.merge(laws, how="outer")


df

```
