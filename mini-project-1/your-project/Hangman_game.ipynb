```python
# Function to draw the Hangman.

import turtle

def Drawing(j):
    t = turtle.Pen()

    #Gallows
    t.forward(200)
    t.left(90)
    t.forward(100)
    t.left(135)
    t.forward(140)
    t.left(135)
    t.forward(100)
    t.left(90)
    t.forward(300)
    t.left(90)
    t.forward(100)
    t.left(90)
    t.forward(40)
    t.up()

#   print(t.position())

# I could not find the way to refer to a specific part of the function, therefore
# with every wrong guess the function redraws the gallows + already done body parts.
# If that's possible, here I could set absolute starting position and probably should set the correct direction (south).

#   t.setposition(101.01, 261.01)

    t.down()
    
    #Head, i = 1
    t.up()
    t.right(90)
    t.down()
    t.circle(20)
    if j == 1:
        return

    #Neck, i = 2
    t.up()
    t.left(90)
    t.forward(40)
    t.down()
    t.forward(20)
    if j == 2:
        return

    #Left arm, i = 3
    t.right(45)
    t.forward(60)
    t.up()
    t.backward(60)
    t.down()
    if j == 3:
        return

    #Right arm, i = 4
    t.left(90)
    t.forward(60)
    t.up()
    t.backward(60)
    t.down()
    if j == 4:
        return

    #Body, i = 5
    t.right(45)
    t.forward(70)
    if j == 5:
        return

    #Left leg, i = 6
    t.right(45)
    t.forward(60)
    t.up()
    t.backward(60)
    t.down()
    if j == 6:
        return

    #Right leg, i = 7
    t.left(90)
    t.forward(60)
    t.up()
    t.backward(60)
    t.down()
    if j == 7:
        return

# There are two live players in this game, therefore the first player's input is hidden.

import getpass

w = getpass.getpass("Type a word for other player to guess: ")
print("The word has %d letters." %(len(w)))

# Turning input into lowercase

word = w.lower()
print("".join(" _ " for l in word))

guessed_letters = ''

# Function for second player's input. Capital letters are turned to lower case.

def UserGuess():
    letter = input("Guess a letter: ")
    letter = letter.lower()
    return letter

# Function for feedback loop and keeping the score.

def FeedbackLoop():
    i = 0
    guessed = guessed_letters
    wrongletters = []
    while i < 7 and (len(guessed) < len(word)):
        let = UserGuess()
        if let in word:
            print("Yey, you guessed right!")
            guessed = guessed + let
            print("".join(l if l in guessed else " _ " for l in word))
        else:
            print("Wrong, guess again.")
            print("".join(l if l in guessed else " _ " for l in word))
            wrongletters.append(let)
            print("Wrong letters you guessed already: ", wrongletters)
            i += 1
            print("Wrong guesses out of 7: ", i)
            # Also draw one body part of a hangman, call the drawing function.
            Drawing(i)
    if i == 7:
        print("You lost. RIP")
    else:
        print("Congratulations, you are alive!")

FeedbackLoop()
```
