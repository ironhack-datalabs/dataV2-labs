## Challenge 1

#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1

plt.plot(x, y, label='linear', linewidth=2.5, color='black')
plt.plot(x, z, label='quadratic', linewidth=2.5, color='green')

plt.xlabel('x label')
plt.ylabel('another label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2

fig, [ax1, ax2] = plt.subplots(1,2)

ax1.plot(x,y, color = 'black', linewidth=1)
ax1.set_title('Linear relationship')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

ax2.plot(x,z, color = 'green', linewidth=1)
ax2.set_title(' Quadratic relationship')
ax2.set_xlabel('x label')
ax2.set_ylabel('z label')

plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here

fig, [ax1, ax2] = plt.subplots(1,2, figsize=(12,5))

ax1.plot(x,y, color = 'black', linewidth=1)
ax1.set_title('Linear relationship', fontsize=15)
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

ax2.plot(x,z, color = 'green', linewidth=1)
ax2.set_title(' Quadratic relationship', fontsize=15)
ax2.set_xlabel('x label')
ax2.set_ylabel('z label')

plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

# So is it in the same plot or in two subplots?

fig, [ax1, ax2] = plt.subplots(1,2, figsize=(12,5))

ax1.plot(x, x**2, color = 'red', linewidth=1)
ax1.set_title('y = x**2', fontsize = 15)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xscale('linear')

ax2.plot(x, np.exp(x), color = 'green', linewidth=1)
ax2.set_title(' y = e**x', fontsize = 15)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_yscale("log")

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# fitbit.groupby(['Months_encoded']).median()['Steps']
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

fig, ax = plt.subplots(1,2, figsize=(12,4))

plt.sca(ax[0])

fitbit.groupby(['Months_encoded']).median()['Steps'].plot(kind = 'bar')
plt.title('Median steps by months', fontsize = 15)
plt.xlabel('Months_encoded')
plt.ylabel('Median steps')
plt.xticks(range(13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

# The fitbiters are more active on weekdays (1 is for workday).

plt.sca(ax[1])

fitbit.groupby(['Work_or_Weekend']).median()['Steps'].plot(kind = 'bar')
plt.title(' Median steps by workday or weekend ', fontsize = 15)
plt.xlabel('Work_or_Weekend')
plt.ylabel('Median steps')
plt.xticks(range(2), ['Weekend', 'Workday'])

plt.savefig('../figures/figures.png')

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

lst = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

f, ax = plt.subplots(1, 3, figsize=(15,4))

for i in range(len(lst)):
    ax[i].scatter(fitbit[lst[i]], fitbit['Steps'])
    ax[i].set_title(lst[i]+' vs Steps')
    ax[i].set_xlabel(lst[i])
    ax[i].set_ylabel('Steps')

plt.show()
```
