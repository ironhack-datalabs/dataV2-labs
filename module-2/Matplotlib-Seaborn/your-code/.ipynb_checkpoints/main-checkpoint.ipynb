#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1
plt.plot(x,y, 'go--', label='linear', linewidth=2, markersize=6)
plt.plot(x,z, label='quadratic', color='blue', marker='x', linestyle='dashed', markersize=4)

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("linear vs quadratic plot")
plt.legend()
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x,y)
ax1.set_title('linear')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

ax2.plot(x,z)
ax2.set_title('quadratric')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(15,7))
ax1.plot(x,y)
ax1.set_title('linear')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label', fontsize='15')

ax2.plot(x,z)
ax2.set_title('quadratric')
ax2.set_xlabel('x label', fontweight='bold')
ax2.set_ylabel('y label')

plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.

**Hint**: Use `set_xscale` and `set_yscale`.

```python
import math

x = np.arange(0,100)
y = x*2
z = np.exp(x)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('x')
ax1.set_ylabel('x', color=color)
lns1 = ax1.plot(x, y, color=color,  marker='o', label = 'square(x)')
ax1.tick_params(axis='y', labelcolor=color)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('exp(x)', color=color)  # we already handled the x-label with ax1
lns2 = ax2.plot(x, z, color=color,  marker='x', label = 'exp(x)')
ax2.set_yscale('log')
ax2.tick_params(axis='y', labelcolor=color)

#Combine 2 labels
lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0, shadow=True)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
