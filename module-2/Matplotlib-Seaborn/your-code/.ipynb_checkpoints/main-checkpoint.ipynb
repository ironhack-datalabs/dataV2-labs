#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig, [ax1,ax2]=plt.subplots(1,2,figsize=(10,10))
ax1.plot(x,x**2)
ax1.set_title('Square')

y = np.exp(x)
ax2.plot(x,y)
ax2.set_title('Exponential')

plt.set_xscale = 'linear'
plt.set_yscale = 'logscale'

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

# fitbit.plot.bar( x='Months_encoded',y='Steps')
fb = fitbit.pivot_table( index =['Months_encoded'],aggfunc = {'Steps':np.median})

#fb.plot.bar()

b = list(fb.Steps)
a = list(fb.index)

fig, [ax1,ax2]=plt.subplots(1,2,figsize=(15,5))
ax1.bar(a,b)
ax1.set_title('Steps per Month')
ax1.set_xlabel('Months')
ax1.set_ylabel('Steps')


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

fb = fitbit.pivot_table( index =['Work_or_Weekend'],aggfunc = {'Steps':np.median})
b = list(fb.Steps)
c = list(fb.index)
ax2.bar(c,b)
ax2.set_title('Workday v Weekend Steps')

#This is my way of Bullying this plot into submission!!!!!!
ax2.set_xticklabels(['    ','    ','Workday','     ','       ','      ','       ','Weekend'])

plt.show()

```

#### Use the right plot to visualize the column `Pclass`.

```python
# your code here
#titanic['Pclass'].hist()

labels = 'Third Class', 'First Class', 'Second Class'
sizes = titanic['Pclass'].value_counts()
explode = (0.1, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels,autopct='%1.1f%%', 
        shadow=True, startangle=90 )
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Titanic Passenger Classes')
plt.show()
```
