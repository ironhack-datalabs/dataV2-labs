#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x_square = np.square(x)
x_exp = np.exp(x)

fig, ax = plt.subplots(nrows=1, ncols =2, figsize=(15,5))
ax[0].plot(x, x_square)
ax[0].plot(x, x_exp)
ax[0].set_title('y= x**2')
ax[0].set_xlabel('x')
ax[0].set_ylabel('echelle classique')
ax[1].plot(x, x_square)
ax[1].plot(x,x_exp)
ax[1].set_title('echelle logarithmique')
ax[1].set_xlabel('x')
ax[1].set_ylabel('e^x')
ax[1].set_yscale("log")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]
fig, ax = plt.subplots(nrows=len(features), ncols =1, figsize=(15,15))

for counter, chart in enumerate(ax):
    current_feature = features[counter]
    chart.scatter(fitbit[current_feature], fitbit["Steps"]) 
    chart.set_title(f'{current_feature} vs Steps')
    chart.set_xlabel(current_feature)
    chart.set_ylabel('Steps')
```

#### Plot the correlation matrix using seaborn.

```python
# Compute the correlation matrix
corr = titanic.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
