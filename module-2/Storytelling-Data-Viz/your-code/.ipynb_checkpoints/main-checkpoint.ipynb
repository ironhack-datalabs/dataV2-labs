```python
# your code here
# add cells as needed
updated_values_df = df[['SalePriceAdjusted','TotalBathrooms','TotalRooms','AgeofHouse','YearsSinceRemodel']]

sns.pairplot(updated_values_df)
```

```python
df['Neighborhood'].unique()
```

```python
uni_1 = df['Neighborhood'].unique()
uni_1

nhood_totals = []

for h in uni_1:
        nhood_totals.append(df[(df['Neighborhood'] == h)]['SalePrice'])
                            
# Your code here
nhood_totals
len(nhood_totals) # you should see 18
```

```python
nhood_avg_price = df.pivot_table(index='Neighborhood',values = 'SalePrice', aggfunc=np.mean)
#nhood_avg_price['SalePrice'].dtype
# list_avg_price = list(nhood_avg_price['SalePrice'])
# list_avg_price
nhood_avg_price
```

```python
#Ho = the average sale prices for the neighborhoods are similar.
#H1 = the averages are significantly different.

stats.f_oneway(*nhood_totals)
```

```python
nhood_avg_price.plot.bar()
```

```python

column_list = ['FullBath','HalfBath','BsmtFullBath', 'BsmtHalfBath']
df['TotalBathrooms']=df[column_list].sum(axis=1)

bath_total = df.pivot_table(index= 'TotalBathrooms',values = 'SalePrice',aggfunc = np.mean)
df.head()
```

```python
col_list = ['BedroomAbvGr', 'KitchenAbvGr', 'FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath']
df['TotalRooms']=df[col_list].sum(axis=1)

total_rooms = df.pivot_table(index = 'TotalRooms',values = 'SalePrice')
df.head()
```

```python
df['AgeofHouse']= 2020 - df['YearBuilt']

age_house = df.pivot_table(index = 'AgeofHouse',values = 'SalePrice')
df.head()
```

```python
df['YearsSinceRemodel'] = 2020 - df['YearRemodAdd']
newly_remodeled = df.pivot_table(index = 'YearsSinceRemodel', values = 'SalePrice')
df.head()
```

## Bonus Challenge 2 - Exploring Data with Correlation Heatmap

Now you have explored data visualizations with certain fields based on common sense. In the dataset there are many other fields that you are not sure whether they are important factors for the sale price. What is the best way to explore those fields without investigating them individually?

Making scatter matrix is not an option here because there are too many fields which makes it extremely time consuming to create scatter matrix. One option you have is to create a heatmap. Heatmaps are much less expensive to create than scatter matrixes. You can use heatmaps to visualize the pairwise correlations between each two variables.

Here is a [reference](https://seaborn.pydata.org/examples/many_pairwise_correlations.html) you can use to learn how to creat the pairwise correlation heatmap. Your heatmap should look like below:

![Corr Heatmap](heatmap.png)

```python
# your code here
#sns.heatmap(df.corr())
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

```

In your heatmap, you can easily identify the highly correlated (either positively or negatively) variables by looking for the grids with darker colors.
