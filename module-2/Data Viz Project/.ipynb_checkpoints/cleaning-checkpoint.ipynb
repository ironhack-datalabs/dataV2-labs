# There are more missing value the first years so here we will only take the last ten years of values with the columns Y2004 to Y2014. There is a lot of column which we will not use, we are looking at the production quantities so we can create a new dataframe with only interesting columns.

```python
crops_last_tenyears = crops[['Area','Item','Element','Unit','Y2004','Y2005',
                                    'Y2006','Y2007','Y2008','Y2009','Y2010','Y2011','Y2012','Y2013','Y2014']]
crops_last_tenyears.info()
```

```python
crops_last_tenyears[crops_last_tenyears.isnull().any(axis=1)]
```

```python
crops_cleaned = crops_last_tenyears.dropna()
crops_cleaned.info()
```

```python
# We will only look at the 'tonnes' unit, other units are duplicates in other way
crops_cleaned[crops_cleaned.Unit == 'tonnes'].to_csv('european_crops_cleaned.csv', index=False)
```
