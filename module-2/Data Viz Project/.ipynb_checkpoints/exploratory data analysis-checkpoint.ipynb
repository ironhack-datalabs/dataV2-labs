```python
import pandas as pd
wheat = pd.read_csv('european_wheat_cleaned.csv')
```

```python
wheat.head()
```

```python
meaned_data = wheat.groupby(['Item']).mean().reset_index()
meaned_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
ploted_data = meaned_data.melt('Item', var_name='cols',  value_name='vals')
ploted_data['cols'] = ploted_data['cols'].apply(lambda x: x[1:]).astype(int)
ploted_data.head()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_yscale('log')
f = sns.lineplot(x="cols", y="vals", hue='Item', 
     data=ploted_data.query(
    '''Item=="Apples" or Item=="Watermelons" or Item=="Artichokes" or Item=="Bananas" or Item=="Cabbages and other brassicas"'''
    ), ax=ax)
```

# Insights: 2005 and 2010 seems like a hard year, almost all Items production decrease. 2007 also but not for Bananas and Artichokes. We must confirm that with more data

```python
ploted_data.groupby(['cols']).sum().plot.line()
```

# We grouped all items by year and sum the production. We can see that our insights were good : 2005, 2007 and 2010 are indeed hard years for production. We also see that 2012 is also a hard year and we didn't saw that on our sample

# Now we want to see the production by country if some countries are more affected than others. Let's focus on 2012 and 2014

```python
meaned_data_bycountry = wheat.groupby(['Area']).sum()
meaned_data_bycountry.head()
```

```python
ploted_data_by_country = meaned_data_bycountry[['Y2012','Y2014']].reset_index().melt('Area', var_name='Year',  value_name='production')
ploted_data_by_country['Year'] = ploted_data_by_country['Year'].apply(lambda x: x[1:]).astype(int)
ploted_data_by_country
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
fig, ax = plt.subplots(figsize=(10,10))
ax = sns.barplot(x="Area", y="production", hue="Year", data=ploted_data_by_country.query(
    '''Area=="Austria" or Area=="France" or Area=="Ukraine" or Area=="United Kingdom" or Area=="Germany"'''
    ))
```

## *We can see that this little sample shows that 2014 is a better year than 2012

```python
meaned_data_bycountry[meaned_data_bycountry.Y2012 > meaned_data_bycountry.Y2014]
```

```python
import numpy as np
meaned_data_bycountry['2012_is_better'] = np.where(meaned_data_bycountry.Y2012 > meaned_data_bycountry.Y2014, 1, 0)
```

```python
country_with_better_2012_y = meaned_data_bycountry['2012_is_better'].sum()

df4 = pd.DataFrame({"2012_is_better":["Yes", "No"], 
                    "BestYear":[country_with_better_2012_y,meaned_data_bycountry.shape[0]-country_with_better_2012_y]})

_, ax = plt.subplots(figsize = (10,10))
wedges,_,_ = ax.pie(df4['BestYear']
                    ,labels=df4["2012_is_better"]
                    ,shadow=False,startangle=90, autopct="%1.1f%%"
                    ,textprops={'fontsize': 16})
ax.set_title("Did you have a better production in 2012 comparing it with 2014 ?")
ax.legend(wedges,df4["2012_is_better"], loc="upper center", prop={'size': 16});
```
