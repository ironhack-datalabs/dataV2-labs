####  What is the average purchase price?

```python
# your code here
print('The average purchase price is: ' + str(orders['UnitPrice'].mean()) + '.')
```

#### What are the highest and lowest purchase prices? 

```python
# your code here
max_price = orders['UnitPrice'].max()
min_price = orders['UnitPrice'].min()
print(f'The highest price is: {max_price}.')
print(f'The lowest price is: {min_price}.')
```

#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
# your code here
orders_spain = orders[orders['Country'] == 'Spain']
orders_spain.head()
```

#### How many customers do we have in Spain?

```python
# your code here
customers_spain = len(set(orders_spain['CustomerID']))
print(f"There are {customers_spain} customers in Spain.")
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here
items_per_customer = orders.groupby('CustomerID')[['Quantity']].sum()
items_50plus_per_customer = items_per_customer[items_per_customer['Quantity'] >50]
items_50plus_per_customer
```

#### Select orders from Spain that include more than 50 items.

```python
# your code here
# Assumption is that every order results in one invoice
orders_spain_grouped = orders_spain.groupby('InvoiceNo')[['Quantity']].sum()
orders_spain_items_50plus = orders_spain_grouped[orders_spain_grouped['Quantity'] >50]
orders_spain_items_50plus
```

#### Select all free orders.

```python
# your code here
orders_free = orders[orders['amount_spent'] == 0]
orders_free
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
# your code here
orders_lunch_bag = orders[orders['Description'].str.startswith('lunch bag')]
orders_lunch_bag
```

#### Select all `lunch bag` orders made in 2011.

```python
# your code here
orders_lunch_bag_year = orders_lunch_bag[orders_lunch_bag['year'] == 2011]
orders_lunch_bag_year
```

#### Select all orders made in the month of August.

#### What's the  average amount of money spent by country?
