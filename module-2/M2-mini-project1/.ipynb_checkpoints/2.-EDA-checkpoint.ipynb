## 1. The `color` column
First, let's look at the color column.

**For each different color, find the mean of each column. You should have a matrix with every color as rows and the columns `carat`, `clarity`, etc as columns.**

```python
#your code here

columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z', 'clarity_nr']

diamonds.groupby(['color'])[columns].agg(['mean'])
```

**What do you see? Pay special attention to the relationship between price and color.**

```python
#your thoughts here

# https://www.bluenile.com/education/diamonds/color
# For reference, the industry standard is as follows:
# D-F colorless diamonds - the rarest and highest quality with a pure icy look. 
# G-J are near-colorless diamonds, great value for the quality.

# I was a bit surprised to see that average price increased for worse colored diamonds.

# The more pure-colored diamonds are smaller in carats and in other mesurements
# (x, y, z).

# On average, the most clear specimens are fall into G and F color category.
```

Let's go further into the color feature. We will plot the frequency distribution of the diamonds color in our dataset. 

**Plot the distribution and analyze it. Remember that you can use the pandas `plot()` method.**

```python
#your code here

diamonds.groupby('color').size().plot(kind='bar')
```

```python
#your comments here

# The frequency distribution is right-skewed, i.e. there are more diamonds
# graded D-G color than others. The most popular is the mid-range, G color. 
```

## 2.The `carat` column 

Let's check the `carat` (weight), since this could also be a potential factor for price change.

**Find the mean of each column for each value of `carat` using the `groupby` method. Then comment your results.**

```python
#your code here

columns = ['depth', 'table', 'price', 'x', 'y', 'z', 'color_nr', 'clarity_nr']

diamonds.groupby(['carat'])[columns].agg(['mean'])
```

```python
#your comments

# The average price siginificantly increases for heavier diamonds, also ones with 
# higher x, y and z measurements.
# The heavier the diamond, the less likely it is going to be highly graded for its 
# color or clarity.
```


**Plot a histogram of the `carat` column by using the `plot` method (see the docs to find an easy way to do so). What is happening?**

```python
#your code here

diamonds['carat'].plot.hist(bins = 100, figsize = (8, 4))
```

```python
# Most diamonds in this dataset weight below 2 carats.
```

# Bonus: taking a deeper look with plots and correlations

To take deeper look, we will use the `pairplot` method of `seaborn` library. This method plots a scatterplot for each pair of features and in the diagonal the distribution of the feature.

So if you have many features it will take a while, be careful!


```python
#Run this code

sns.pairplot(diamonds, diag_kind = 'kde', plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'})
```

**What do you see here? What relationships between variables are the most interesting?**

```python
#your thoughts here

# As observed before, there is positive relationship between price and carat, i.e. diamond
# weight. Naturally, x, y and z measurements increase with weight and show the same
# pattern paired with price in the scatter plot. 
# Moreover, clarity and color do not show any significant relationship with weight, price or
# other measurements.
# Carat and price distributions are indeed right-skewed, as predicted analysing the mean and 
# meadian above, i.e. most diamonds in this dataset are lighter and more affordable.
```

Now we will see a correlation matrix with a plot. As you know a higher correlation means that the feature could be an effect (**but is not for sure**) for the changes on the price.

We will see this with a matrix with colors. A lighter color means greater correlation. 

This is done with the `seaborn` library as well.

```python
#Run this code
plt.figure(figsize=(8, 8))
p = sns.heatmap(diamonds.corr(), annot=True, square=True)
```

**What do you see here? Regarding the results before, does it fit with them?**

```python
#your thoughts here

# The correlations between price and carat, price and x, y, z are very strong. 
```

Finally, we will calculate the linear regression between the price and the weight. This will be done first by plotting it with the `seaborn` library and then calculating the error with the `scipy` library.

```python
#Run this code
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(diamonds.carat, diamonds.price)
r2 = r_value ** 2
r2
```

```python
#Run this code
plt.figure(figsize=(5, 5))
sns.regplot(diamonds.carat, diamonds.price, scatter=True)
```

**What do you think?**

```python
#your thoughts here

# 85% of the price change is explained by change in carat.
```
