```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("CDD.csv", sep = "\t")
data
```

```python
data.corr()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,10)) 
heatmap = sns.heatmap(data.corr(), annot=True)
```

```python
data["ESTIMATED TOTAL COST"].corr(data["FATALITIES"])
#these two variables are almost not correlated at all just because there is no real relation between the 
#estimated total costs and the fatalities for some reason.
```

```python
data["INJURED / INFECTED"].corr(data["FATALITIES"])
#obviously there is a high correlation between injured/infected people and total fatalities...
```

```python
data["OGD COSTS"].corr(data["UTILITY - PEOPLE AFFECTED"])
#these two seem to be slightly negatively correlated. 
#ogd costs go down if utility-people affected goes up and vice versa
```

```python
#printing all
pairplot = sns.pairplot(data)
```

```python
data_reduced1 = data[["ESTIMATED TOTAL COST","FATALITIES"]]
pairplot2 = sns.pairplot(data_reduced1)
#this is more or less what one could have expected. The scatterplot indicate that there is no correlation 
```

```python
data_reduced2 = data[["INJURED / INFECTED","FATALITIES"]]
pairplot2 = sns.pairplot(data_reduced2)
#there is a very big outlier in this case making it difficult to see a really good relationship
#in my last example (at the bottom of this python file one can find a nicer scatterplot)
```

```python
data_reduced3 = data[["OGD COSTS","UTILITY - PEOPLE AFFECTED"]]
pairplot3 = sns.pairplot(data_reduced3)
#this is more or less what one could have expected, due to the very low negative correlation, the scatterplot 
#actually seems to indicate no relationship at all
```
