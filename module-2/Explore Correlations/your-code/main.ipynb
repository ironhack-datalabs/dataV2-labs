```python
can_dis = pd.read_csv('../data/CDD.csv', sep='\t')
can_dis
```

```python
can_dis.corr()
```

```python
import seaborn as sns
%matplotlib inline
sns.heatmap(can_dis.corr())
```

```python
#1) Highly positively correlated pair: UTILITY - PEOPLE AFFECTED and NGO PAYMENTS. 
#We see this positive correlation between the amount of people whose utility services 
#were interrupted/affected by a specific event and NGO payments provided by the government.
can_dis['INJURED / INFECTED'].corr(can_dis['NGO PAYMENTS'])
```

```python
#2) Correlation is almost zero pair:
#There is no correlation between the magnitude (other than earthquake) and a fatality. 
#This is because "Measure" values are for an earthquake and there are very little to 
#none earthquakes in Canada.
A measure of the size of an earthquake, related to the amount of energy released.

can_dis['MAGNITUDE'].corr(can_dis['FATALITIES'])
```

```python
#3) Negatively correlated pair:
#There is very little negative correlation between the amount of people whose utility services 
#were interrupted/affected and OGD Costs.
can_dis['UTILITY - PEOPLE AFFECTED'].corr(can_dis['OGD COSTS'])
```
