```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/vpavandijk/LEARNING/Ironhack/copy of google drive class AMS0520/Copy of CDD.csv', sep = '\t')
data.head()
```

```python
#Print the correlation matrix:

data.corr()
```

```python
#Print a heat map of this correlation matrix (hint: use seaborn.heatmap())

data.corr().style.background_gradient(cmap='coolwarm')
```

```python
# Print a heat map of this correlation matrix (hint: use seaborn.heatmap())

# sns.heatmap(data.corr())

fig, ax = plt.subplots(figsize=(18,18))
sns.heatmap(data.corr(),annot=True, linewidths=.5, ax=ax)
```
